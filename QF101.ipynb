## Solving ODE numerically


### Q1. Apply explicit Euler's method with step size h=1/4 to the following initial value problem on the interval \[0,1\]
<br>

$\frac {dy} {dt} = \frac 1 {y^2}$ , $y(t_0 = 0)$

```python
y = 1
step_size = 0.25
current_step = 0 
explicit_euler_y_values = []
interval = [0,1000]
while current_step < interval[1]:
    explicit_euler_y_values.append(y)
    y = y + step_size*(1/(y**2))
    current_step += step_size

print(explicit_euler_y_values[:4])
plt.figure(figsize=(15,6))
plt.plot(explicit_euler_y_values)
plt.title("Q1 Euler Method")
print('Range extended to {} for visualisation purposes'.format(interval[1]))
plt.show()
```


### Q2. Apply two step Euler's method with step size h=1/4 to the following initial value problem

$y_1^{'} = - y_1 - y_2$ , $y_1(t = 0) = 1$

$y_2^{'} = y_1 - y_2$ , $y_2(t = 0) = 0$

```python
y_1 = 1
y_2 = 0
step_size = 0.25
current_step = 0
interval = [0,1000]
y_1_values = []
y_2_values = []
while current_step < interval[1]:
    y_1_values.append(y_1)
    y_2_values.append(y_2)
    temp_y1 = y_1
    temp_y2 = y_2 
    y_1 = y_1+ step_size*(-temp_y1-temp_y2)
    y_2 = y_2 + step_size*(temp_y1-temp_y2)
    current_step += step_size
print(y_1_values[:4])
print(y_2_values[:4])
plt.figure(figsize=(15,6))
plt.plot(y_1_values)
plt.title("Q2 y_1 function")
print('Range extended to {} for visualisation purposes'.format(interval[1]))

plt.figure(figsize=(15,6))
plt.plot(y_2_values)
plt.title("Q2 y_2 function")
plt.show()
```

### Q4. Apply two step Implicit Euler's method with step size h=1/4 to the following initial value problem

$\frac {dy} {dt} = 10(1-y)$ , $y(t_0=0)=0$

$y(t_1)$ = $y(t_0=0) + h*(10(1-y(t_1)))$

Rearranging the equation, we get

$y(t_1) = \frac {y(t_0) + 10h} {1+10h}$

```python
y = 0
step_size = 0.25
current_step = 0
implicit_euler_y_values = []
interval = [0,100]
while current_step < interval[1]:
    implicit_euler_y_values.append(y)
    y = (y+10*step_size)/(1+10*step_size)
    current_step += step_size

print(implicit_euler_y_values[:4])
plt.figure(figsize=(15,6))
plt.plot(implicit_euler_y_values)
plt.title("Q4 Implicit Euler Method")
print('Range extended to {} for visualisation purposes'.format(interval[1]))
plt.show()
```

### Q5. Apply RK method with step size h=1/4 to the following initial value problem

$\frac {dy} {dt} = \frac 1 {y^2}$ , $y(t_0 = 0)$

```python
y = 1
step_size = 0.25
current_step = 0
rk_y_values = []
interval = [0,1000]
while current_step < interval[1]:
    rk_y_values.append(y)
    # Using the standard h/6(1S1 + 2S2 + 2S3 + S4) RK ratio
    s1 = 1/ (y**2)
    s2 = 1/ ((y + (step_size/2)*s1) **2)
    s3 = 1/ ((y + (step_size/2)*s2) **2)
    s4 = 1/ ((y +(step_size*s3)) **2 )
    y = y + (step_size/6)*(s1 + 2*s2 + 2*s3 + s4)
    current_step += step_size

print(rk_y_values[:4])
plt.figure(figsize=(15,6))
plt.plot(rk_y_values)
plt.title("Q5 RK Method")
print('Range extended to {} for visualisation purposes'.format(interval[1]))
plt.show()
```

## Q5 Part 2. Compare the error between RK method and Explicit Euler Method

```python
rk_vector = np.array(rk_y_values)
euler_vector = np.array(explicit_euler_y_values)
error = rk_vector-euler_vector

plt.figure(figsize=(15,6))
plt.plot(explicit_euler_y_values)
plt.plot(rk_y_values)
plt.title("Q5 Part 2 - RK vs. Euler")

plt.figure(figsize=(15,6))
plt.plot(error)
plt.title("Q5 Part 2 - RK vs. Euler , Error size")
plt.show()
```

# Pesudo Random Number Generator

Using the industry standard of linear congruence Generator - generate samples between \[0 , 1\]

```python
a = 16087 # This is the amplification factor
seed = 1
sample_size = 1000
period = (2**32)- 1 
vals = np.zeros(sample_size)
x = seed
for i in range(sample_size):
    x = (a*x) % period 
    vals[i] = x
adjusted_vals = vals / period 

plt.figure(figsize=(15,15))
plt.scatter([i for i in range(sample_size)],adjusted_vals)
plt.title('Pseudo Random Number Generator')
plt.show()
```
